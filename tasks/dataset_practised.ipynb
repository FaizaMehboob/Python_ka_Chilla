```python
a = pd.read_csv("data2.csv")
sns.set_theme(style="ticks", color_codes=True)
g=sns.FacetGrid(a,row="Element", hue="Year")
g=(g.map(plt.scatter,"Element","Year").add_legend())

g.set_titles()

plt.show()

```

```python
import seaborn as sns
sns.set_theme()

a=pd.read_csv('data2.csv')

# Plot sepal width as a function of sepal_length across days
g = sns.lmplot(
    data=a,
    x="Year", y="Value", hue="Item Code",
    height=5
)

# Use more informative axis labels than are provided by default
g.set_axis_labels("Years", "Values")
```
