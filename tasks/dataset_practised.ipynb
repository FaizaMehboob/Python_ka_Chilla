```python
a = pd.read_csv("data1.csv")
sns.set_theme(style="ticks", color_codes=True)
g=sns.FacetGrid(a,row="Element", hue="Year")
g=(g.map(plt.scatter,"Element","Year").add_legend())

g.set_titles()

plt.show()

```

```python
a = pd.read_csv("data4.csv")

sns.set_theme(style="ticks", color_codes=True)
p=sns.countplot(x="Value", data=a)
p.set_title("New_one Graph")
plt.show
```

```python
a = pd.read_csv("data4.csv")
sns.pairplot(a, hue="Year")

p.set_title("New_one Graph")

plt.show
```

```python
a = pd.read_csv("data4.csv")

sns.lineplot(x="Year", y="Item Code", data=a)
plt.title("Graph")
#plt.xlim(2)
#plt.xlim(3)
plt.show()
```

```python
a = pd.read_csv("data4.csv")
sns.lineplot(x="Year", y="Element Code", data=a)
sns.lineplot(data=a)
plt.title("darkgrid")
plt.show()
```

```python
a = pd.read_csv("data4.csv")
sns.set(style='whitegrid')

sns.boxplot(x="Area", y="Value",data=a)
```
